

# Introducing StaPy - (automatic) Static Traffic Assignment in Python


```python
#importing packages

from network_data import get_from_ox_and_save, load_pickle
from assignment_methods import DUE
from od_matrix_generator import generate_od, generate_od_fixed
from visualization import plot_network, show_desire_lines

```

```python
# first we download all the data from osm, apply some filtering, estimate capacity and generate a routeable graph

g, deleted= get_from_ox_and_save('Gent')
```

```python
plot_network(deleted, mode='deleted elements', notebook=True) 
```

```python
#this is simply generating an origin destination table with a given amount of origin destination pairs


rand_od = generate_od_fixed(g.number_of_nodes(), number_of_od_values=20 )
```

```python
#desire lines illustrate the randomly generated demand pattern

show_desire_lines(g, rand_od, notebook=True)
```

```python
#next we run the actual assignment by use of the Method of Successive Averages (MSA),
#delays are estimated based on Volume Delay Functions
#

DUE(g, od_matrix=rand_od, method='bpr,flow_avg')
```

```python
plot_network(g, notebook=True)
```

```python
from visualization import show_convergence
show_convergence(g, notebook=True)
```

```python
# we can also plot individual iterations of the algorithm
plot_network(g, notebook=True, iteration=6)
```
